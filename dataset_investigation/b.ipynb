```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
df = pd.read_csv('source_data/csv/healthcare_expenditure.csv')
```

```python
df.head()
```

```python
# replace spaces with underscores and make lowercase
df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
```

```python
df.head()
```

```python
# rename country column in place
df.rename(columns={'per_capita_total_expenditure_on_health_(ppp_int._$)': 'country'}, inplace=True)
df.head()
```

```python
# pull out just the Canada data
dfc = df.loc[df['country'] == 'Canada']
```

```python
dfc.head()
```

```python
# drop country column
cols = [0]
dfc.drop(dfc.columns[cols], axis=1, inplace=True)
dfc.head()
```

```python
# transpose it
dfc = dfc.T
dfc.head()
```

```python
# rename the columns
dfc.index.name = 'year'
dfc.columns = ['value']
dfc.head()
```

```python
# drop if any value in the row has a nan
dfc.dropna(how='any', inplace = True);
```

```python
# for plotting ease, make the year a column
dfc.reset_index(inplace = True)
dfc.head()
```

```python
dfc.plot();
```

```python
plt.plot(dfc['year'], dfc['value'])

plt.title('Per capita total expenditure on health, Canada')
plt.xlabel('Year')
plt.ylabel('PPP int. $')

plt.show()
```
